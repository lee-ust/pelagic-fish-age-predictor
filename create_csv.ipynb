```python
#use environmental attribute and physical trait to predict age of species
#load the data
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

df = pd.read_csv('D:/program/aquatic/data/aquatic_animal.csv')
print(df.columns)
selected_df = df[['Common_Name','Habitat_Type', 'Obs_Depth_m',
                 'Body_Length_cm','Body_Weight_kg', 'Sex',
                 'Water_Temp_C', 'Salinity_ppt','pH','Estimated_Age_yr']]
```

```python
selected_df['Common_Name'].unique()
```

```python
#select typical pelagic species
pelagic_species = [
    'Albacore Tuna',
    'Yellowfin Tuna',
    'Bluefin Tuna',
    'Swordfish',
    'Mahi-Mahi',
    'Blue Marlin',
    'Wahoo',
    'Sailfish'
]

pelagic_species_df = selected_df[selected_df['Common_Name'].isin(pelagic_species)]
```

```python
pelagic_species_df.describe()
```

```python
#filter unknown sex
pelagic_species_df = pelagic_species_df[pelagic_species_df['Sex'] != 'Unknown']
pelagic_species_df['Sex'].unique()
```

```python
#save the data
pelagic_train_set,pelagic_test_set = train_test_split(pelagic_species_df, test_size=0.3, random_state=42)
pelagic_train_set.to_csv('D:/program/aquatic/data/pelagic_training_set.csv',index=False)
pelagic_test_set.to_csv('D:/program/aquatic/data/pelagic_test_set.csv',index=False)
```
